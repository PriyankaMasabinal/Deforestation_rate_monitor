```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("deforestation_cleaned.xlsx")
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap="coolwarm",
            fmt=".2f")

plt.title("Correlation Matrix")
plt.show()
```

```python
# Agricultural Expansion vs Forest Loss
# ===============================
plt.figure(figsize=(7,5))
sns.scatterplot(
    x="agricultural_expansion_sqkm",
    y="forest_loss_sqkm",
    data=df
)

plt.title("Agricultural Expansion vs Forest Loss")
plt.show()
```

```python
# Cattle Density vs Forest Loss
# ===============================
plt.figure(figsize=(7,5))
sns.scatterplot(
    x="cattle_density",
    y="forest_loss_sqkm",
    data=df
)

plt.title("Cattle Density vs Forest Loss")
plt.show()
```

```python
# Forest Loss
# ===============================
plt.figure(figsize=(7,5))
sns.histplot(df["forest_loss_sqkm"], kde=True)

plt.title("Distribution of Forest Loss")
plt.show()
```

```python
# 7. Forest Loss Trend Over Years
# ===============================
plt.figure(figsize=(10,5))

yearly_loss = df.groupby("year")["forest_loss_sqkm"].mean()

plt.plot(yearly_loss.index, yearly_loss.values)
plt.title("Average Forest Loss Over Time")
plt.xlabel("Year")
plt.ylabel("Forest Loss (sqkm)")
plt.show()
```
