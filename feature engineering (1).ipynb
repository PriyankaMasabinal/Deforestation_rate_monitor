```python
import pandas as pd
```

```python
df = pd.read_excel("deforestation_cleaned.xlsx")
```

```python
# 3. Agricultural Pressure Index
# ===============================
# Measures how aggressively agriculture expands relative to forest size

df["agri_pressure_index"] = (
    df["agricultural_expansion_sqkm"] /
    df["forest_area_sqkm"]
)
```

```python
# 3. Agricultural Pressure Index
# ===============================
# Measures how aggressively agriculture expands relative to forest size

df["agri_pressure_index"] = (
    df["agricultural_expansion_sqkm"] /
    df["forest_area_sqkm"]
)
```

```python
# 5. Combined Land Pressure (KEY FEATURE ⭐)
# ===============================
# Captures acceleration effect (interaction term)

df["combined_land_pressure"] = (
    df["agricultural_expansion_sqkm"] *
    df["cattle_density"]
)
```

```python
# 6. Deforestation Intensity
# ===============================
# Forest loss relative to available forest

df["deforestation_intensity"] = (
    df["forest_loss_sqkm"] /
    df["forest_area_sqkm"]
)
```

```python
# 7. Replace infinite values (safety step)
# ===============================
df.replace([float("inf"), -float("inf")], pd.NA, inplace=True)

# ===============================
# 8. Save Feature Engineered Dataset
# ===============================
df.to_excel("feature_engineered_dataset.xlsx", index=False)

print("Feature Engineering Completed Successfully!")
```
